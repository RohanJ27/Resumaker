```python
import os
import PyPDF2
import pandas as pd

def extract_pdf_text(file_path):
    text = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        
        for page_number in range(num_pages):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()
    
    return text

# Folder path containing the PDF files
folder_path = "archive/data/data"


df = pd.DataFrame()
# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    pdf_texts = []
    if filename!=".DS_Store":
        subfolder_path = os.path.join(folder_path, filename)
        for filenames in os.listdir(subfolder_path):
            file_path = os.path.join(subfolder_path, filenames)
            extracted_text = extract_pdf_text(file_path)
            pdf_texts.append(extracted_text)
        df[filename] = pdf_texts
        


# Print the DataFrame
print(df)
```
